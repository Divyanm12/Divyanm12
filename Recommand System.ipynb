```python
import pandas as pd
import numpy as np
books = pd.read_csv("book.csv")
print(books.shape) 
books.columns
```

```python
books.head()
```

```python
books = books.rename(columns={'User.ID':'userid','Book.Title':'book','Book.Rating':'rating'})
books.head()
```

```python
df=pd.DataFrame(books)
```

```python
#number of unique users
len(books.UserID.unique())
```

```python
#number of unique books
len(books.BookTitle.unique())
```

```python
book_df = books.pivot_table(index = 'UserID',columns = 'BookTitle',values ='BookRating').reset_index(drop=True)
book_df.index = books.UserID.unique()
book_df
```

```python
#Impute those NaNs with 0 values
book_df.fillna(0,inplace = True)
book_df
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
user_sim = 1 - pairwise_distances(book_df.values,metric = 'cosine')
user_sim
```

```python
np.fill_diagonal(user_sim, 0)
user_sim
```

```python
#store it in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books.UserID.unique()
user_sim_df.columns = books.UserID.unique()
user_sim_df
```

```python
def get_user_similar_books(user1,user2):
    common_books = books[books.UserID == user1].merge(books[books.UserID == user2],
                                                     on = 'BookTitle',
                                                     how = 'inner')
    return common_books.merge(books, on = 'BookTitle')
```
